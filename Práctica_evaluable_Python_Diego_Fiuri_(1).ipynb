```python
def total_compra(unidades:int, precio_funda= 9.5, descuento= 0.25):

  total_bruto= precio_funda * unidades
  total_neto= total_bruto - (descuento * total_bruto)

  if unidades >= 3:
    total_final= total_neto
  else:
    total_final= total_bruto

  return round(total_final,2)

print("El total será finalmente de:", total_compra(2
                                                   ))

```

9. Discretización de la edad

```python
# Como ya existia la col "age_class" se creó la col "age_class2" a traves de una función con IF y al final agregando en una columna el resultado para cada valor de "Age" del dataframe

def clasificar_edad(Age):

  if Age < 21:
     return "Muy Joven"

  if 21 <= Age <= 26:
    return "Joven"

  if 27 <= Age <= 32:
    return "Veterano"

  else:

    return "Muy veterano"

df["age_class2"] = df["Age"].apply(clasificar_edad)

display(df)
```

10. Diagrama de barras

```python
import matplotlib.pyplot as plt

#Primer intento

barplot_data = df.age_class.value_counts()
barplot_data= barplot_data.rename_axis("unique_values").reset_index(name="counts")
barplot_data.unique_values = barplot_data.unique_values.astype(str)

fig, ax = plt.subplots()
ax.bar(barplot_data.unique_values, barplot_data.counts, color="green")
plt.title("Players by Age class")
plt.xlabel("Age Class")
plt.ylabel("Sum of Players")

plt.show()


```
